```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.airlinequality.com/review-pages/a-z-airport-reviews/'
response = requests.get(url)
content = response.text
```

```python
#print(content)
soup = BeautifulSoup(content, 'lxml')
#print(soup)
Reviews = []

review = soup.find_all('div', class_ = 'text_content')
#print(reviews)
for i in range(1,1001):
    for i in review:
        main_review = i.text
        Reviews.append(main_review.replace('\n', ''))
print(Reviews)
```

```python
df = pd.DataFrame()
df["Reviews"] = Reviews
df.head(15)
```

```python
df.to_csv("reviews.csv")
```
